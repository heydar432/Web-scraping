```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
from bs4 import BeautifulSoup
import requests
import numpy as np
```

```python
url = 'https://en.wikipedia.org/wiki/2022_FIFA_World_Cup'
res = requests.get(url)
print(res)
content = res.text
soup = BeautifulSoup(content, 'lxml')
print(soup)
```

```python
# correct method

tables = soup.find_all('table', class_ = 'fevent')
tables

yenisi = [] 
for j in tables:
    home_cons = []
    for i in j.find_all('a'):
        title = i.text
        home_cons.append(title)
    print(home_cons)
    yenisi.append(home_cons)
```

```python
df = pd.DataFrame(yenisi).iloc[:,:4]
df.columns = ['team1', 'ft_result', 'team_2', 'yokk']
df
```

```python
for i in df.index:
    for j in range(len(df.columns)):
        if df.iloc[i,j] == 'a.e.t.':
            df.iloc[i,j] = df.iloc[i,j+1]

del df['yokk']

df
```
