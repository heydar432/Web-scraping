```python
# import libraries for project

from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import time
import pandas as pd
import numpy as np
```

```python
# Base URLs
BASE_URL = 'https://www.trendyol.com/sr?q=blue+tooth+kulak%C3%BCst%C3%BC+kulakl%C4%B1k&qt=blue+tooth+kulak%C3%BCst%C3%BC+kulakl%C4%B1k&st=blue+tooth+kulak%C3%BCst%C3%BC+kulakl%C4%B1k&os=1&pi='
PRODUCT_BASE_URL = 'https://www.trendyol.com'

# Setting up the browser
driver = webdriver.Chrome()

products_data = []

for page_num in range(208):  # Adjust range as needed

    print('Page :',page_num)

    url = BASE_URL + str(page_num)
    driver.get(url)
    time.sleep(2 + np.random.rand())

    main_soup = BeautifulSoup(driver.page_source, 'lxml')
    link_divs = main_soup.find_all('div', class_='p-card-chldrn-cntnr card-border')
    time.sleep(np.random.rand() * 3)

    for i in link_divs:
        anchor_tag = i.find('a')
        if anchor_tag:
            product_url = PRODUCT_BASE_URL + anchor_tag['href']
            driver.get(product_url)
            time.sleep(2 + np.random.rand())

            product_soup = BeautifulSoup(driver.page_source, 'lxml')
            data_dict = {'product_link': product_url}
            
            data_combinations = [
                ('span', 'prc-dsc'),
                ('div', 'rating-line-count'),
                ('span', 'total-review-count'),
                ('span', 'favorite-count'),
                ('div', 'campaign-name'),
                ('span', 'dd-txt-vl'),
                ('div', 'sl-pn'),
                ('a', 'seller-name-text')
            ]

            for tag_type, class_name in data_combinations:
                tag = product_soup.find(tag_type, class_=class_name)
                data_dict[class_name] = tag.text.strip() if tag else 'Not found'

            attributes = product_soup.find_all('li', class_='attribute-item')
            attributes_dict = {}
            unique_features = ["Çift Telefon Desteği", "Dokunmatik Kontrol", "Bluetooth Versiyon", "Renk", "Garanti Süresi", "Aktif Gürültü Önleme (ANC)", "Mikrofon", "Suya/Tere Dayanıklılık", "Model", "Garanti Tipi"]

            for attr in attributes:
                label = attr.find('span', class_='attribute-label')
                value = attr.find('span', class_='attribute-value')
                if label and value and label.text.strip() in unique_features:
                    attributes_dict[label.text.strip()] = value.text.strip()

            combined_data = {**data_dict, **attributes_dict}
            products_data.append(combined_data)

            print(combined_data)
            time.sleep(np.random.rand() * 0.3)

df = pd.DataFrame(products_data)

# Load the existing data
file_name = "products_data.xlsx"
with pd.ExcelFile(file_name) as xls:
    if 'Sheet1' in xls.sheet_names:
        existing_data = pd.read_excel(xls, 'Sheet1')
    else:
        existing_data = pd.DataFrame()

# Concatenate the new data to the existing data
combined_data_last = pd.concat([existing_data, df], ignore_index=True)

# Save the combined data back to the Excel file
with pd.ExcelWriter(file_name, engine='openpyxl', mode='w') as writer:
    combined_data_last.to_excel(writer, index=False)

driver.quit()
```
