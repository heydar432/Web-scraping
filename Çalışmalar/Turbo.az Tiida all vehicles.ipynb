```python
url1 = 'https://turbo.az/autos?'
pages = 'page='
tails = '&q%5Bbarter%5D=0&q%5Bcrashed%5D=1&q%5Bcurrency%5D=azn&q%5Bengine_volume_from%5D=&q%5Bengine_volume_to%5D=&q%5Bfor_spare_parts%5D=0&q%5Bloan%5D=0&q%5Bmake%5D%5B%5D=7&q%5Bmileage_from%5D=&q%5Bmileage_to%5D=&q%5Bmodel%5D%5B%5D=&q%5Bmodel%5D%5B%5D=485&q%5Bonly_shops%5D=&q%5Bpainted%5D=1&q%5Bpower_from%5D=&q%5Bpower_to%5D=&q%5Bprice_from%5D=&q%5Bprice_to%5D=&q%5Bregion%5D%5B%5D=&q%5Bsort%5D=price_asc&q%5Bused%5D=&q%5Byear_from%5D=&q%5Byear_to%5D='
yenisi = []

for j in range(1,14):


    res = requests.get(url1 + pages + str(j) + tails)
    content = res.text
    soup = BeautifulSoup(content, 'lxml')

    class_kred = soup.find_all('div', class_ = 'products-i__info')
    kred = []
    for i in class_kred :
        title = i.text
        kred.append(title)

    print(kred)
    class_price = soup.find_all('div', class_ = 'product-price')
    price = []
    for i in class_price :
        title = i.text
        price.append(title)

    print(price)
    class_marka = soup.find_all('div', class_="products-i__name products-i__bottom-text")
    markası = []
    for i in class_marka:
        title = i.text
        markası.append(title)

    print(markası)
    class_ilmotorkm = soup.find_all('div', class_="products-i__attributes products-i__bottom-text")
    ilmotorkm = []
    for i in class_ilmotorkm:
        title = i.text
        ilmotorkm.append(title)

    print(ilmotorkm)
    class_city = soup.find_all('div', class_="products-i__datetime")
    city = []
    for i in class_city:
        title = i.text
        city.append(title)

    print(city)

    yenisi.extend(zip(kred, price, markası, ilmotorkm, city))
```

```python
def min_max_scaling(data, feature_range=(0, 1)):
    min_val, max_val = feature_range
    return (data - np.min(data)) / (np.max(data) - np.min(data)) * (max_val - min_val) + min_val
```

```python
for l in df[['year','price(AZN)','totald(km)']]:
    df[l+'_scale'] = min_max_scaling(df[l])
```

```python
df['car_score'] = df['year_scale'] - df['price(AZN)_scale'] - df['totald(km)_scale']
```
